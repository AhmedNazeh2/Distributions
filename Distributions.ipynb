```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, binom, poisson, expon, uniform, lognorm

# Set a common seed for reproducibility
np.random.seed(42)

# Create a figure with 3 rows and 2 columns
fig, axs = plt.subplots(3, 2, figsize=(12, 12))

# 1. Normal Distribution (Adult Heights in cm)
heights_data = np.random.normal(loc=170, scale=10, size=1000)
axs[0, 0].hist(heights_data, bins=50, density=True, color='blue', alpha=0.7)

x = np.linspace(140, 200, 1000)
axs[0, 0].plot(x, norm.pdf(x, loc=170, scale=10), 'r-', lw=2)

axs[0, 0].set_title('Adult Heights (Normal Distribution)')

# 2. Binomial Distribution (Number of heads in 10 coin flips)
n, p = 10, 0.5
coin_flips_data = np.random.binomial(n=n, p=p, size=1000)
axs[0, 1].hist(coin_flips_data, bins=20, density=True, color='green', alpha=0.7)

x = np.arange(0, 11)
axs[0, 1].plot(x, binom.pmf(x, n, p), 'r-', lw=2)

axs[0, 1].set_title('Coin Flips (Binomial Distribution)')

# 3. Poisson Distribution (Number of emails received per hour)
lam = 5
emails_data = np.random.poisson(lam=lam, size=1000)
axs[1, 0].hist(emails_data, bins=20, density=True, color='orange', alpha=0.7)

x = np.arange(0, 20)
axs[1, 0].plot(x, poisson.pmf(x, lam), 'r-', lw=2)

axs[1, 0].set_title('Emails Received per Hour (Poisson Distribution)')

# 4. Exponential Distribution (Time between arrivals of customers in minutes)
scale = 10
arrivals_data = np.random.exponential(scale=scale, size=1000)
axs[1, 1].hist(arrivals_data, bins=50, density=True, color='red', alpha=0.7)

x = np.linspace(0, 100, 1000)
axs[1, 1].plot(x, expon.pdf(x, scale=scale), 'r-', lw=2)

axs[1, 1].set_title('Time Between Arrivals (Exponential Distribution)')

# 5. Uniform Distribution (Random number generator between 0 and 1)
uniform_data = np.random.uniform(low=0, high=1, size=1000)
axs[2, 0].hist(uniform_data, bins=50, density=True, color='purple', alpha=0.7)

x = np.linspace(0, 1, 1000)
axs[2, 0].plot(x, uniform.pdf(x, loc=0, scale=1), 'r-', lw=2)

axs[2, 0].set_title('Random Number Generator (Uniform Distribution)')

# 6. Lognormal Distribution (Income distribution in thousands of dollars)
lognormal_data = np.random.lognormal(mean=3, sigma=0.5, size=1000)
axs[2, 1].hist(lognormal_data, bins=50, density=True, color='brown', alpha=0.7)

x = np.linspace(0, max(lognormal_data), 1000)
axs[2, 1].plot(x, lognorm.pdf(x, s=0.5, scale=np.exp(3)), 'r-', lw=2)

axs[2, 1].set_title('Income Distribution (Lognormal Distribution)')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
